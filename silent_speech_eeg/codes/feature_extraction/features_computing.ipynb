# Markers time stamps

```python
txt_dir = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/markers'
txt_paths = glob.glob(txt_dir+'/*.txt')
output_files = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/time_stamps/'
```

```python
for path in txt_paths:
    name = path.split('/')[-1].split('.')[0]
    start_list = []
    stop_list = []
    with open(path,'r') as txt:
        for index,line in enumerate(txt):
            if index%2 == 0:
                start_list.append(round(float(line.split(',')[2])/5))
            else:
                stop_list.append(round(float(line.split(',')[2])/5))
    time_stamps = pd.DataFrame({'start':start_list,'stop':stop_list})
    time_stamps.to_csv(output_files+name+'.csv',index=False)
```

# Trimming

```python
cwd = os.getcwd()
eeg_dir = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/eeg/'
mfcc_dir = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/mfcc/'
time_dir = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/time_stamps/'
eeg_paths = glob.glob(eeg_dir+'/*.csv')
time_paths = glob.glob(time_dir+'/*.csv')
mfcc_paths = glob.glob(mfcc_dir+'/*csv')
mfcc_paths.sort()
eeg_paths.sort()
time_paths.sort()
output = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/trimmed_eeg/'
```

```python
def trimm_markers(eeg_df,time_df):
    eeg_df['in'] = 0
    for row in range(len(time_df)):
        start,stop = time_df.iloc[row][['start','stop']]
        eeg_df.loc[start:stop,'in'] = row+1
    eeg_df = eeg_df[eeg_df['in']!=0].reset_index(drop=True)
    return eeg_df
```

```python
for eeg_path,time_path in zip(eeg_paths,time_paths):
    name = eeg_path.split('/')[-1].split('.')[0]
    eeg = pd.read_csv(eeg_path)
    time = pd.read_csv(time_path)
    eeg = trimm_markers(eeg,time)
    eeg.to_csv(output+name+'.csv',index=False)
```

```python
for eeg_path,time_path in zip(mfcc_paths,time_paths):
    name = eeg_path.split('/')[-1].split('.')[0]
    eeg = pd.read_csv(eeg_path)
    time = pd.read_csv(time_path)
    eeg = trimm_markers(eeg,time)
    eeg.to_csv(eeg_path,index=False)
```

# Dimension reduction

```python
cwd = os.getcwd()
csv_path = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/trimmed_eeg/'
csv_paths = glob.glob(csv_path+'/*.csv')
output_path = '/home/guatam-admin/Database/first_30_sen_background_music_with_markers/dimension_reduced_eeg/'
```

```python
for path in csv_paths:
    name = path.split('/')[-1].split('.')[0]
    df = pd.read_csv(path)
    df_label = df[df.columns[-1]]
    X = df[df.columns[:-1]].values
    scaler = StandardScaler()
    X = scaler.fit_transform(X)
    kpca = KernelPCA(n_components=30, kernel='poly')
    df = kpca.fit_transform(X)
    df = pd.DataFrame(df)
    df['sentence'] = df_label
    df.to_csv(output_path + name+'.csv',index=False)
    print(name+'...done....')
```
